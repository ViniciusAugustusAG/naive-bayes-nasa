Este modelo prevê se um objeto próximo à Terra é ou não perigoso. <br>
Base de dados: NASA & Nearest Earth Objects


```python
#Para importar os dados
import pandas as pd

#Para Naive Bayes
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics

#Para dividir conjunto de treinamento e teste
from sklearn.model_selection import train_test_split
```

```python
#Visualizando formato do dataset
data = pd.read_csv("drive/My Drive/Colab Notebooks/TEIA/Dataset/neo.csv")
data.shape
```

```python
data.head()
```

```python
#Observando os dados e checando dados faltantes
data.info()
```

```python
#Limpando dados e dividindo x e y
x = data.drop(["hazardous","id","name",'est_diameter_max',"est_diameter_max","miss_distance","absolute_magnitude","orbiting_body","sentry_object"] , axis = 1)

y = data.hazardous.astype("int")
print(data.shape, y.shape)
```

```python
#Dividindo conjuntos de treinamento e teste
X_train, X_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 0)
```

```python
#Aplicando o Modelo de Gauss Naive Bayes
GNB = GaussianNB()

GNB.fit(X_train, y_train)

y_pred = GNB.predict(X_test)

print("Acurácia:",metrics.accuracy_score(y_test, y_pred))
```
